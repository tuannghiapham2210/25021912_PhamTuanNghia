```python
#Bài 6: Nhập vào số nguyên dương a, in ra tất cả các ước dương của a
#xử lý hàm
def uoc(n):
    uoc = [i for i in range(1, n + 1) if n % i == 0]
    return uoc

#xử lý input
while True:
    try:
        n = int(input("hãy nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("hãy nhập vào một số nguyên dương: "))
        break
    except ValueError:
        pass

#xử lý output
for i in uoc(n):
    if i > 0:
        print(i, end = " ")
```

```python
#Bài 7: Nhập vào số nguyên dương a, đếm số ước dương của a
#xử lý hàm
def dem_uoc(n):
    lst = [i for i in range (1, n + 1) if n % i == 0]
    return len(lst)

#xử lý input
while True:
    try:
        n = int(input("hãy nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("hãy nhâp vào một số nguyên dương: "))
        break
    except ValueError:
        pass

#xử lý output
print(dem_uoc(n))
```

```python
#Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b
#xử lý hàm
def uoc_chung(a, b):
    uoc_a = [i for i in range(1, a + 1) if a % i == 0]
    uoc_b = [i for i in range(1, b + 1) if b % i == 0]
    uoc_chung = [i for i in uoc_a if i in uoc_b]
    return uoc_chung

#xử lý input
while True:
    try:
        a, b = map(int, input("nhập vào hai số nguyên dương (hai số ngăn cách nhau bởi khoảng trắng): ").split())
        while a <= 0 or b <= 0:
            a, b = map(int, input("nhập vào hai số nguyên dương (hai số ngăn cách nhau bởi khoảng trắng): ").split())
        break

    except ValueError:
        pass

#xử lý output
print("các ước chung của 2 số là:", end= " ")
for i in uoc_chung(a, b):
    print(i, end = " ")
```

```python
#Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….
import math
#xử lý hàm
def chinhphuong(n):
    if math.sqrt(n).is_integer():
        return True
    return False

#xử lý input
while True:
    try:
        n = int(input("hãy nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("hãy nhập vào một số nguyên dương: "))
        break

    except ValueError:
        pass

#xử lý output
print(f"{n} là một số chính phương" if chinhphuong(n) else f"{n} KHÔNG phải một số chính phương")
```

```python
#Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…
#xử lý hàm
def perfect_number(n):
    lst = [i for i in range(1, n) if n % i == 0]
    if sum(lst) == n:
        return True
    return False

#xử lý input
while True:
    try:
        n = int(input("hãy nhập vào đây một số nguyên dương: "))
        while n <= 0:
            n = int(input("hãy nhập vào đây một số nguyên dương: "))
        break

    except ValueError:
        pass

#xử lý output
print(f"{n} là số hoàn hảo" if perfect_number(n) else f"{n} không phải số hoàn hảo")
```

```python
#Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n
#xử lý hàm
def find_k(n):
    Sk = 0
    for i in range(1, n + 1):
        Sk += i
        if Sk >= n:
            return i - 1

#xử lý input
while True:
    try:
        n = int(input("Nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("Nhập vào một số nguyên dương: "))
        break

    except ValueError:
        pass

#xử lý output
print(find_k(n))
```

```python
#Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A
#xử lý hàm
def find_n(a):
    tong = 0
    dem = 1
    while tong <= a:
        tong += 1/dem
        dem += 1
    return dem - 1

#xử lý input
while True:
    try:
        a = int(input("hãy nhập vào một số nguyên dương: "))
        while a <= 0:
            a = int(input("hãy nhập vào một số nguyên dương: "))
        break
    except ValueError:
        pass

#xử lý output
print(find_n(a))
```

```python
#Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng
try:
    n = int(input("Hãy nhập vào đây một số nguyên (nhập -1 để thoát): "))

except ValueError:
    print("Giá trị của bạn không hợp lệ!")
    exit()

if n == -1:
    print("Bạn chưa nhập gì cả :(")

else:

    lon_nhat = n
    be_nhat = n

    while True:
        try:
            n = int(input("Hãy nhập vào đây một số nguyên (nhập -1 để thoát): "))
        except ValueError:
            print("Giá trị của bạn không hợp lệ!")
            exit()

        if n == -1:
            break
        if n < be_nhat:
            be_nhat = n
        if n > lon_nhat:
            lon_nhat = n

    print(f"Số bé nhất là: {be_nhat}", f"Số lớn nhất là: {lon_nhat}", sep = "\n")
```

```python
#Bài 16: hập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ
#xử lý hàm
def chan_le(n):
    chan = 0
    le = 0
    for i in str(n):
        if int(i) % 2 == 0:
            chan += 1
        else:
            le += 1
    return chan, le

#xử lý input
while True:
    try:
        n = int(input("Hãy nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("Hãy nhập vào một số nguyên dương: "))
        break
    except ValueError:
        pass

#xử lý output
print(f"Số chữ số chẵn là: {chan_le(n)[0]}", f"Số chữ số lẻ là: {chan_le(n)[1]}", sep = "\n")
```

```python
#Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không
#xử lý hàm
def ham(n):
    while n % 3 == 0:
        n //= 3
    if n == 1:
        return True
    else:
        return False

#xử lý input
while True:
    try:
        n = int(input("Nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("Nhập vào một số nguyên dương: "))
        break
    except ValueError:
        pass

#xử lý output
if ham(n):
    print("Yes")
else:
    print("No")
```

```python
#Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A.
#xử lý hàm
def fibonacci(n):
    lst = [1, 1]
    while True:
        lst.append(lst[-1] + lst[-2])
        if lst[-1] > n:
            break
    return lst[-2]

#xử lý input
while True:
    try:
        n = int(input("Hãy nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("Hãy nhập vào một số nguyên dương: "))
        break
    except ValueError:
        pass

#xử lý output:
print(fibonacci(n))
```

```python
#Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.
#xử lý hàm
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def prime_counter(n):
    total = 0
    for i in str(n):
        if is_prime(int(i)):
            total += 1
    return total

#xử lý input
while True:
    try:
        n = int(input("Nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("Nhập vào một số nguyên dương: "))
        break
    except ValueError:
        pass

#xử lý output
print(prime_counter(n))
```
