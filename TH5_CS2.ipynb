```python
#W5A1
def Max_Of_Two(a , b):
    if a > b:
        return a
    elif b > a:
        return b
    else:
        return a
```

```python
#W5A2
def Swap_Numbers(a, b):
    return tuple((b, a))
```

```python
#W5A3
import math
def Prime_Number(a):
    if a < 2:
        return False
    for i in range (2, int(math.sqrt(a) + 1)):
        if a % i == 0:
            return False
    return True
```

```python
#W5A4
def Perfect_Number(a):
    s = 0
    for i in range(1, a):
        if a % i == 0:
            s+=i
    if s == a:
        return True
    else:
        return False
```

```python
#W5A5
def Location(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
```

```python
#W5A6
def GiaiThua(n):
    if n == 0:
        return 1
    return n * (n - 1)
```

```python
#W5A7
def Calculator(a, b, c):
    if c == "-":
        return a - b
    elif c == "+":
        return a + b
    elif c == "*":
        return a * b
    elif c == "/":
        return a / b
```

```python
#W5A8
def Hamming(a, b):
    x = a ^ b
    cnt = 0
    while x > 0:
        if x & 1 == 1:
            cnt += 1
        x = x >> 1
    return cnt
```

```python
#W5A9
def Tong(n):
    return n * (n + 1) / 2
```

```python
#W5A10
def DangCau(a, b):
    if len(a) != len(b):
        return False
    dic = {}
    myset = set()
    for i, j in zip(a, b):
        if i in dic:
            if dic[i] != j:
                return False
        else:
            if j in myset:
                return False
            dic[i] = j
            myset.add(j)
    return True
```
