```python
#W7A16

from collections import Counter

def beautiful_array(arr):
    freq = Counter(arr)
    cnt = 0
    for num, fre in freq.items():
        if fre > num:
            cnt += fre - num
        elif fre < num:
            cnt += fre
    return cnt
            
arr = map(int, input().split())

res = beautiful_array(arr)

print(f"NO {res}" if res != 0 else "YES")
```

```python
#W7A17 (Boyer Moore Majority Vote Algorithm)

arr = list(map(int, input().split()))

n = len(arr)

candidate = None
count = 0

for i in range(n):
    if count == 0:
        candidate = arr[i]
        count = 1
    elif candidate == arr[i]:
        count += 1
    else:
        count -= 1
    
count = 0
for i in range(n):
    if candidate == arr[i]:
        count += 1

print(f"{candidate}" if count > n / 2 else "") 
    
```

```python
#W7A19

a1 = list(map(int, input().split()))
a2 = list(map(int, input().split()))

n1 = len(a1)
n2 = len(a2)

res = []

if a1[0] <= a1[-1]:
    p1 = p2 = 0
    while p1 < n1 and p2 < n2:
        
        if a1[p1] <= a2[p2]:
            res.append(a1[p1])
            p1 += 1
        
        else:
            res.append(a2[p2])
            p2 += 1

    if p1 < n1:
        res.extend(a1[p1:])

    elif p2 < n2:
        res.extend(a2[p2:])

else:
    p1 = p2 = 0
    while p1 < n1 and p2 < n2:
        
        if a1[p1] >= a2[p2]:
            res.append(a1[p1])
            p1 += 1
        
        else:
            res.append(a2[p2])
            p2 += 1

    if p1 < n1:
        res.extend(a1[p1:])

    elif p2 < n2:
        res.extend(a2[p2:])

print(*res)
```

```python
#W7A25

num = input()

k = int(input())

remove = len(num) - k

stack = []

for digit in num:
    while stack and remove > 0 and digit > stack[-1]:
        stack.pop()
        remove -= 1

    stack.append(digit)

if remove > 0:
    stack = stack[0:k]

print("".join(stack))
```

```python
#W7A27
string = input()

#quy tắc của dãy ngoặc đúng là LIFO (last in first out), tức là dấu ngoặc nào mở cuối cùng thì phải đóng đầu tiên
#ví dụ: ([{}]), rõ ràng, { mở đầu cuối nhưng phải đóng đầu tiên --> sử dụng stack để check
#([{}
def verify(string):
    mapping = {"}" : "{", "]" : "[", ")" : "(", ">" : "<"}

    stack = []

    for char in string:
        #nếu là ngoặc đòng
        if char in mapping:
            #nếu stack rỗng (không có ngoặc mở trước đó) hoặc ngoặc mở trước đó không khớp --> False
            if not stack or stack.pop() != mapping[char]:
                return False
        #nếu là ngoặc mở
        else:
            stack.append(char)
    #nếu stack rỗng, thì các ngoặc đã được ghép với nhau --> True
    #nếu stack không rỗng --> thừa ra ngoặc --> False
    return True if not stack else False


res = verify(string)

print("YES" if res else "NO")
```

```python
#W7A28 (tìm dãy các hợp số dài nhất liên tiếp < n)
import math

n = int(input())

is_prime = [True] * (n + 1)

def sieve():
    #loại bỏ thủ công 0 và 1 (không phải số nguyên tố)
    is_prime[0] = is_prime[1] = False
    #duyệt từ 2 đến căn n + 1
    for i in range(2, math.isqrt(n + 1) + 1):
        #nếu số nào là số nguyên tố, loại bỏ tất cả các bội của nó
        if is_prime[i]:
            for j in range(i*i, n + 1, i):
                is_prime[j] = False

sieve()

current_sequence = []

longest_sequence = []

#số 4 là hợp số đầu tiên nên ta duyệt từ 4
for i in range(4, n):
    if not is_prime[i]:
       current_sequence.append(i)

    else:
        if len(current_sequence) > len(longest_sequence):
            longest_sequence = current_sequence
            current_sequence = []      

        else:
            current_sequence = []

if len(current_sequence) > len(longest_sequence):
    longest_sequence = current_sequence

print(*longest_sequence, sep = ", ")
```

```python
#W7A29
nums = list(map(int, input().split()))

current_sequence = [nums[0]]

longest_sequence = []

for i in range(1, len(nums)):
    if nums[i] > nums[i - 1]:
        current_sequence.append(nums[i])
    
    else:
        if len(current_sequence) > len(longest_sequence):
            longest_sequence = current_sequence
            current_sequence = [nums[i]]
        
        else:
            current_sequence = [nums[i]]

if len(current_sequence) > len(longest_sequence):
    longest_sequence = current_sequence

print(*longest_sequence) 
```

```python
#W7A30
nums = list(map(int, input().split()))

n = len(nums)

#ban đầu, độ dài tất cả các dãy con tăng dần tại mọi vị trí đều bằng 1
L = [1] * n 

P = [-1] * n

for i in range(n):
    for j in range(i):
        if nums[i] >= nums[j] and L[j] + 1 > L[i]:
            L[i] = L[j] + 1
            #lưu vết: số tại vị trí i được nối vào số tại vị trí j
            P[i] = j

#truy vết:
res = []

#tìm index đầu tiên mà tại đó độ dài dãy con là dài nhất
curr = L.index(max(L))

while curr != -1:
    res.append(nums[curr])
    curr = P[curr]

print(*res[::-1])

```

```python
#W7A31
string = input()

num = [char for char in string]

n = len(string)

for i in range(n - 1):
    swapped = False
    min_idx = i
    for j in range(i + 1, n):
        if num[j] < num[i] and num[j] <= num[min_idx]:
            min_idx = j
    if min_idx != i:
        num[i], num[min_idx] = num[min_idx], num[i]
        swapped = True
        break

if not swapped:
    for i in range(n - 1):
        similar = False
        for j in range(i + 1, n):
            if num[j] == num[i]:
                similar = True
    
    if similar:
        print("".join(num))
    
    else:
        num[-1], num[-2] = num[-2], num[-1]
        print("".join(num))

else:
    print(int("".join(num)))
```

```python
#W7A31 (dùng hàm để tối ưu)

def solve():
    num = list(input().strip())
    n = len(num)

    for i in range(n - 1):
        min_idx = i

        for j in range(i, n):
            if num[j] < num[i] and num[j] <= num[min_idx]:
                min_idx = j

        if min_idx != i:
            num[min_idx], num[i] = num[i], num[min_idx]
            return int("".join(num))
        
    #trường hợp đã được sort tăng dần (nhưng có phần tử trùng nhau) vd:123345
    if len(set(num)) < len(num):
        return int("".join(num))
    
    #trường hợp đã được sort tăng dần (nhưng ko có trùng lặp)
    else:
        num[-1], num[-2] = num[-2], num[-1]
        return int("".join(num))

print(solve())
```
