```python
#W6A19
def find_max(lst):
    largest = 0
    for i in range(1, len(lst)):
        if lst[i] > lst[largest]:
            largest = i
    return largest

lst = list(map(int, input().split()))

print(find_max(lst))
```

```python
#W6A20
def find_num(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
        
lst = list(map(int, input().split()))

k = int(input())

print(find_num(lst, k))
```

```python
#W6A21
from collections import defaultdict

list_string = input().split()

dct = defaultdict(list)

for index, string in enumerate(list_string, 0):
    dct[string].append(index)

for key, value in dct.items():
    dct[key] = tuple(value)

print(dict(dct))
```

```python
#W6A23
def delete_smaller(lst, x):
    return [i for i in lst if i >= x]

lst = map(int, input().split())
x = int(input())

print(delete_smaller(lst, x))


```

```python
#W6A28 (idea from gemini but code 100% from me (no help from AI in terms of codes))
lst = list(map(int, input().split()))

#zip lst với lst[1:] sẽ tạo ra các cặp tuple các số liền kề
#vd 1 2 3 4 sẽ được zip với 2 3 4 để tạo ra [(1, 2), (2, 3), (3, 4)]

#nếu tất cả các phần tử bên phải đều lớn hơn các phần tử bên trái thì all() sẽ trả về True (dãy tăng) 
increasing = all(b > a for a, b in zip(lst, lst[1:]))

#ngược lại thì là giảm
decreasing = all(b < a for a, b in zip(lst, lst[1:]))

if increasing:
    print("increasing")
elif decreasing:
    print("decreasing")
else:
    print("anonymous")
```
