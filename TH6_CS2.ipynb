```python
#W6A1
n = map(int, input().split())

cnt = {}

for i in n:
    if i in cnt:
        cnt[i] += 1
    else:
        cnt[i] = 1

print(list(i for i in cnt))
```

```python
#W6A4
from collections import defaultdict

n = input().split()

dct = defaultdict(list)

for i in n:
    key, value = i.split(":")
    dct[key].append(value)

print(dict(dct))
```

```python
#W6A5
import ast

n = input()
my_list = ast.literal_eval(n)

dct = {"positives" : 0, "negatives" : 0, "zeros" : 0}

for i in my_list:
    if i > 0:
        dct["positives"] += 1
    elif i < 0:
        dct["negatives"] += 1
    else:
        dct["zeros"] += 1

print(dct)

```

```python
#W6A9
import ast

dct1 = ast.literal_eval(input())
dct2 = ast.literal_eval(input())

all_keys = set(dct1.keys()) | set(dct2.keys())
new_dct = {key : dct1.get(key, 0) + dct2.get(key, 0) for key in all_keys}

print(dict(sorted(new_dct.items())))
```

```python
#W6A10
import ast

n = ast.literal_eval(input())
k = int(input())

appeared = set()
res = set()

for i in n:
    couple = (k - i)
    if (k - i) in appeared:
        res.add(tuple((couple, i)))
    appeared.add(i)

print(sorted(res))
```

```python
#W6A12
import ast
from collections import defaultdict

n = ast.literal_eval(input())
cnt = defaultdict(int)

for i in n:
    cnt[i] += 1

max_fre = max(cnt.values())

print(min(i for i in cnt.keys() if cnt[i] == max_fre))
```
