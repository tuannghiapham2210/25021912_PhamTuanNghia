```python
#W9A1 
def merge(arr1, arr2):
    i = j = 0
    n1 = len(arr1)
    n2= len(arr2)
    res = []
    
    while i < n1 and j < n2:
        if arr1[i] <= arr2[j]:
            res.append(arr1[i])
            i += 1
        else:
            res.append(arr2[j])
            j += 1
        
    if i < n1:
        res.extend(arr1[i:])

    elif j < n2:
        res.extend(arr2[j:])

    return res

arr1 = list(map(int, input().split()))

arr2 = list(map(int, input().split()))

print(*merge(arr1, arr2))
```

```python
#W9A2 
def flatten(matrix):
    res = []
    for row in matrix:
        if isinstance(row, list):
            res.extend(flatten(row))
        else:
            res.append(row)
    return res

n = int(input())
matrix = [list(map(int, input().split())) for _ in range(n)]
k = int(input())

flattened = flatten(matrix)

flattened.sort()

print(flattened[k - 1])

```

```python
#W9A3 
def swap_minimum(n: int) -> int:
    n = str(n)
    nums = [char for char in n]
    n = len(nums)
    
    #nếu mảng chỉ có 1 phần tử thì return luôn
    if n < 2:
        return int("".join(nums))

    for i in range(n):
        min_idx = i
        
        for j in range(i + 1, n):
            # duyệt từ i+ 1 đến n để tìm phần tử nhỏ nhất và xa nhất
            if nums[j] <= nums[min_idx]:
                min_idx = j
       
        if min_idx != i and nums[i] > nums[min_idx]:
            nums[min_idx], nums[i] = nums[i], nums[min_idx]
            return int("".join(nums))
    

    #nếu mảng được sắp xếp tăng dần và có 2 số giống nhau vd: 1 2 2 3 4, thì swap 2 với 2, tương đương với ko làm gì
    if len(set(nums)) < len(nums):
        return int("".join(nums))


    #trường hợp đặc biệt: nếu số được sắp xếp tăng dần như kiểu 1 2 3 4 5 thì sẽ không swap được phần tử nào ở dưới lên đầu
    # --> swap 2 số cuối
    nums[-1], nums[-2] = nums[-2], nums[-1]
    return int("".join(nums))

n = int(input())
print(swap_minimum(n))

```

```python
#W9A5
def is_subarray(arr1: list, arr2: list) -> bool:
    n = len(arr1)
    i = 0
    sub_array = arr2[:]
    while sub_array != []:
        sub_array = arr2[i: n + i]
        if sub_array == arr1:
            return True
        i += 1
    return False

arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))

res = is_subarray(arr1, arr2)

print("YES" if res else "NO")
```

```python
#W9A6
def next_permutation(arr: list) -> list:
    i = len(arr) - 2

    while i >= 0 and arr[i] >= arr[i + 1]:
        i -= 1
    
    #i mà lớn hơn hoặc = 0 thì vẫn ở trong mảng, chưa phải cấu hình cúi
    if i >= 0:
        j = len(arr) - 1
        while arr[i] >= arr[j]:
            j -= 1
        
        arr[j], arr[i] = arr[i], arr[j]

    arr[i + 1:] = reversed(arr[i + 1:])

    return arr

n = int(input())

arr = list(map(int, input().split()))

print(*next_permutation(arr))
```

```python
#W9A7
def canPlaceFlowers(flowerbed: list, k: int) -> bool:
    #padding thêm 2 số 0 đằng sau đỡ phải if else trường hợp nằm rìa
    check = [0] + flowerbed + [0]
    for i in range(1, len(flowerbed) + 1):
        #nếu cả 2 bên và phần giữa đều trống thì trồng được
        if check[i] == 0 and check[i - 1] == 0 and check[i + 1] == 0:
            check[i] = 1
            k -= 1
    #hết vòng lặp, nếu số cây cần trồng mà nhỏ hơn hoặc bằng 0 --> trồng hết
    return k <= 0

flowerbed = list(map(int, input().split()))
k = int(input())

print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A8
#set lưu tọa độ đã đi qua
myset = [(0, 0)]

moves = input()

#ban đầu, robot ở gốc tọa độ
x = y = 0
for char in moves:
    if char == "R":
        x += 1
        myset.append((x, y))
    elif char == "L":
        x -= 1
        myset.append((x, y))
    elif char == "U":
        y += 1
        myset.append((x, y))
    elif char == "D":
        y -= 1
        myset.append((x, y))

if len(set(myset)) < len(myset):
    print(True)
elif len(set(myset)) == len(myset):
    print(False)


```

```python
#W9A10
import bisect

def min_radius(houses: list, heaters: list) -> int:
    heaters.sort()

    max_radius = 0

    for house in houses:
        pos = bisect.bisect_left(heaters, house)
        left_heater = float("inf")

        if pos > 0:
            left_heater = house - heaters[pos - 1]

        right_heater = float("inf")
        if pos < len(heaters):
            right_heater = heaters[pos] - house

        min_radius = min(left_heater, right_heater)

        max_radius = max(max_radius, min_radius)

    return max_radius


houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))

res = min_radius(houses, heaters)

print(res)
```
