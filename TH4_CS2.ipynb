```python
#W4A1: Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n. (n <= 1000)
#xử lý hàm
def tinhtong(n):
    tong = 0
    for i in range (1, n + 1):
        tong += i
    return tong

#xử lý input
while True:
    try:
        n = int(input())
        while n < 1 or n > 1000:
            n = int(input())
        break

    except ValueError:
        print("hãy nhập vào một số nguyên hợp lệ (từ 1 đến 10): ", end = "")

#xử lý output
print(tinhtong(n))
```

```python
#W4A3: Viết chương trình tính giai thừa của một số nguyên n cho trước.
#xử lý hàm
def giaithua(n):
    result = 1
    for i in range (1, n + 1):
        result *= i
    return result

#xử lý input
while True:
    try:
        n = int(input())
        while n < 0:
            n = int(input("hãy nhập vào một số nguyên dương: "))
        break
    except ValueError:
        print("hãy nhập một số hợp lệ:", end = " ")

#xử lý output
print(giaithua(n))
```

```python
#W4A4: Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu) của số đó. Không dùng kiểu dữ liệu string.
#xử lý hàm
def counter(n):
    dem = 0
    if n < 0:
        n = abs(n)
    while True:
        n = n // 10
        dem += 1
        if n == 0:
            break
    return dem

#xử lý input
while True:
    try:
        n = int(input())
        break

    except ValueError:
        print("hãy nhập vào một số hợp lệ: ", end="")

#xử lý output
print(counter(n))
```

```python
#W4A5: Hãy viết một chương trình nhận vào một số nguyên dương n và một dãy gồm n số nguyên. Nếu n số nguyên đó có số 42 thì in ra màn hình dòng chữ "I′ve found the meaning of life!", ngược lại in ra dòng chữ "It′s a joke!".
#xử lý hàm
def find_42(lst):
    for i in lst:
        if i == 42:
            return "I've found the meaning of life!"
    return "It's a joke!"

#xử lý input
while True:
    try:
        n = int(input())
        lst = []
        for i in range(n):
            lst.append(int(input()))
        break

    except ValueError:
        print("Please try again:", end=" ")

#xử lý output
print(find_42(lst))
```

```python
#W4A6: Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b).
#xử lý hàm
def nguyento(n):
    if n < 2:
        return False
    for i in range (2, n):
        if n % i == 0:
            return False
    return True

#xử lý input
while True:
    try:
        a, b = map(int, input().split())
        while a > b:
            print("hãy nhập vào một khoảng hợp lệ:", end = " ")
            a, b = map(int, input().split())
        break

    except ValueError:
        print("Please try again:", end=" ")

#xử lý output
print(sum(i for i in range(a, b + 1) if nguyento(i)))
```

```python
#W4A7: Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.
#xử lý hàm
def uoc(n):
    lst = []
    for i in range (1, n + 1):
        if n % i == 0:
            lst.append(i)
    return lst

def nguyento(n):
    if n < 2:
        return False
    for i in range (2, n):
        if n % i == 0:
            return False
    return True

#xử lý input
while True:
    try:
        n = int(input("nhập vào một số nguyên dương lớn hơn hoặc bằng 2: "))
        while n < 2:
            n = int(input("nhập vào một số nguyên dương lớn hơn hoặc bằng 2: "))
        break
    except ValueError:
        pass

#xử lý output
print(max(i for i in uoc(n) if nguyento(i)))
```

```python
#W4A8: Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.
#xử lý hàm
def is_palindrome(n):
    n = str(n)
    length = 0
    for _ in n:
        length += 1
    if length % 2 == 0:
        for i in range(int(length / 2)):
            if n[0 + i] != n[-1-i]:
                return False
        return True
    else:
        for i in range(int((length - 1) / 2)):
            if n[0 + i] != n[-1-i]:
                return False
        return True


def palindrome(n):
    count = 0
    while not is_palindrome(n):
        n = str(n)
        reverse_n = n[::-1]
        n = int(n) + int(reverse_n)
        count += 1
    return n, count


#xử lý input
while True:
    try:
        n = int(input("hãy nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("hãy nhập vào một số nguyên dương: "))
        break

    except ValueError:
        pass


#xử lý output
a, b = palindrome(n)
print(f"số bước là: {b}", f"giá trị palindrome thu được là: {a}", sep = "\n")
```

```python
#W4A9: Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).
import math
#xử lý hàm
def chinhphuong(n):
    lst = []
    for i in range(0, n + 1):
        if str(math.sqrt(i)).split(".")[1] == "0":
            lst.append(i)
    return lst

def trunglap(n):
    lst = []
    tong = 0
    for i in str(n):
        lst.append(i)
        tong += 1
    if len(set(lst)) < tong:
        return True
    else:
        return False

#xử lý input
while True:
    try:
        n = int(input("hãy nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("hãy nhập vào một số nguyên dương: "))
        break

    except:
        pass


#xử lý output
for i in chinhphuong(n):
    if not trunglap(i):
        print(i, end = " ")
```

```python
#W4A10: Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất."
#xử lý hàm
def Collazt(n):
    lst = []
    while n != 1:
        if n % 2 == 0:
            lst.append(n)
            n = n / 2
        else:
            lst.append(n)
            n = 3 * n + 1
    lst.append(1)
    return lst

def longest_Collazt(n):
    longest = 0
    for i in range(1, n + 1):
        if len(Collazt(i)) > longest:
            longest = len(Collazt(i))
    lst = []
    for i in range(1, n + 1):
        if len(Collazt(i)) == longest:
            lst.append(i)
    return longest, lst

#xử lý input
while True:
    try:
        n = int(input("nhập vào đây một số nguyên dương: "))
        while n <= 0:
            n = int(input("Nhập vào đây một số nguyên dương"))
        break

    except:
        pass

#xử lý output
a, b = longest_Collazt(n)
print(min(b), a)
```

```python
#W4A11: Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.
#xử lý hàm
def uoc_chan(n):
    lst = []
    for i in range (1, n + 1):
        if n % i == 0:
            if i % 2 == 0:
                lst.append(i)
    return len(lst)


#xử lý input
while True:
    try:
        n = int(input("hãy nhập vào một số nguyên dương: "))
        while n <= 0:
            n = int(input("hãy nhập vào đây một số nguyên dương: "))
        break

    except ValueError:
        pass


#xử lý output
print(uoc_chan(n))
```

```python
#W4A12: Một người có tài khoản tiết kiệm ở ngân hàng và gửi vào X đồng với lãi suất là 0.7% mỗi tháng. Viết chương trình tính số tiền sau N tháng người ấy rút được (cả gốc và lãi, bỏ qua phần lẻ thập phân).
import math
#xử lý hàm
def total(x, n):
    return x * pow((1 + 0.007), n)

#xử lý input
while True:
    try:
        x = float(input("nhập số tiền ban đầu: "))
        n = int(input("nhập số tháng gửi: "))
        while n < 0 or x < 0:
            x = float(input())
            n = int(input())
        break
    except ValueError:
        pass

#xử lý output:
print("số tiền thu được của bạn là:", math.trunc(total(x, n)))
```

```python
#W4A13: Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false".
#xử lý hàm
def close_pair(x, y):
    if x == y:
        return "false"
    uoc_x = []
    for i in range (1, x):
        if x % i == 0:
            uoc_x.append(i)
    uoc_y = []
    for i in range (1, y):
        if y % i == 0:
            uoc_y.append(i)
    if sum(uoc_x) == y and sum(uoc_y) == x:
        return "true"
    else:
        return "false"

#xử lý input
while True:
    try:
        x = int(input("hãy nhập vào số nguyên dương thứ 1: "))
        y = int(input("hãy nhập vào số nguyên dương thứ 2: "))
        while x <= 0 or y <= 0:
            x = int(input("hãy nhập vào số nguyên dương thứ 1: "))
            y = int(input("hãy nhập vào số nguyên dương thứ 2: "))
        break

    except ValueError:
        pass

#xử lý output
print(close_pair(x, y))
```

```python
#W4A14: Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n.
#xử lý hàm:
def uoc_chung_max(x, y):
    uoc_x = [i for i in range(1, x + 1) if x % i == 0]
    uoc_y = [i for i in range(1, y + 1) if y % i == 0]
    uoc_chung = [x for x in uoc_x if x in uoc_y]
    return max(uoc_chung)

#xử lý input
while True:
    try:
        x = int(input("nhập vào số nguyên dương thứ 1: "))
        y = int(input("nhập vào số nguyên dương thứ 2: "))
        while x <= 0 or y <= 0:
            x = int(input("nhập vào số nguyên dương thứ 1: "))
            y = int(input("nhập vào số nguyên dương thứ 2: "))
        break

    except ValueError:
        pass

#xử lý output
print("ước chung lớn nhất của hai số là:", uoc_chung_max(x, y))
```

```python
#W4A15: Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và số lượng chó. Ngược lại, in ra "invalid".
#xử lý hàm
def linear_solver(m, n):
    dogs = (n - 2 * m)/2
    chickens = (4 * m - n)/2
    if str(dogs).split(".")[1] == "0" and str(chickens).split(".")[1] == "0":
        if dogs >= 0 and chickens >= 0:
            return int(chickens), int(dogs)
    return "Invalid"

#xử lý input
while True:
    try:
        m = int(input("hãy nhập vào đây tổng số con gà và chó: "))
        n = int(input("hãy nhập vào đây tổng số chân: "))
        while m < 0 or n < 0:
            m = int(input("hãy nhập vào đây tổng số con gà và chó: "))
            n = int(input("hãy nhập vào đây tổng số chân: "))
        break

    except ValueError:
        pass

#xử lý output
result = linear_solver(m, n)
if result == "Invalid":
    print("Invalid")
else:
    chickens, dogs = result
    print(f"số gà là: {chickens}", f"số chó là: {dogs}", sep="\n")
```
