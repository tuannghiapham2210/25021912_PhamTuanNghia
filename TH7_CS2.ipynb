```python
#W7A1

def binary_search(arr, left, right, target):
    if left > right:
        return -1
    
    mid = (left + right) // 2
    
    if arr[mid] == target:
        return mid
    
    elif arr[mid] > target:
        return binary_search(arr, left, mid - 1, target)
    
    else:
        return binary_search(arr, mid + 1, right, target)


arr = list(map(int, input().split()))

target = int(input())

print(binary_search(arr, 0, len(arr) - 1, target))
```

```python
#W7A2

def count_occurrences(arr, x):
    cnt = 0
    for i in arr:
        if i == x:
            cnt += 1
    return cnt

arr = map(int, input().split())

x = int(input())

print(count_occurrences(arr, x))
```

```python
#W7A3 (Bubble Sort)

# 2 4 3 5 9 6 7 8
# len = 8

def BubbleSort(arr, n):
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if arr[j + 1] < arr[j]:
                arr[j + 1], arr[j] = arr[j], arr[j + 1]
                swapped = True
        if not swapped:
            break

arr = list(map(int, input().split()))

BubbleSort(arr, len(arr))

print(*arr)
```

```python
#W7A3 (Selection Sort)

def SelectionSort(arr, n):
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]
    
arr = list(map(int, input().split()))

SelectionSort(arr, len(arr))

print(*arr)
```

```python
#W7A4
from collections import defaultdict

def largest_freq(arr):
    
    freq = defaultdict(int)

    for i in arr:
        freq[i] += 1
    
    return max(freq.items(), key = lambda x : x[1])

arr = map(int, input().split())

res = largest_freq(arr)

print(f"{res[0]} xuat hien nhieu nhat, som nhat, {res[1]} lan")

```

```python
#W7A5 (làm trâu, 2 vòng for)
def pairs_counter(arr, x):
    n = len(arr)
    cnt = 0
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == x:
                cnt += 1
    return cnt


arr = list(map(int, input().split()))

x = int(input())

print(pairs_counter(arr, x))
```

```python
#W7A5 (dùng dictionary)
from collections import defaultdict

def pairs_counter(arr, x):
    cnt = 0
    seen = defaultdict(int)
    for i in arr:
        mate = x - i
        cnt += seen[mate]
        seen[i] += 1
    return cnt

arr = map(int, input().split())

x = int(input())

print(pairs_counter(arr, x))
```

```python
#W7A6

def flatten(nested_lists):
    res = []
    for i in nested_lists:
        if isinstance(i, list):
            #nếu phần tử đang xét mà cũng là một list thì ta gọi đệ quy để làm phẳng nó
            sub_res = flatten(i)

            #flatten(i) sẽ trả về list con đã được làm phẳng, sau đó ta nối với list cha
            res.extend(sub_res)
            
        #nếu phần tử đang xét ko phải một list (chỉ là một phần tử) thì append vào res luôn
        else:
            res.append(i)

    return res

nested_lists = eval(input())

print(flatten(nested_lists))
```

```python
#W7A7

def longest_sequence(arr):
    n = len(arr)

    #L[i] là độ dài dãy con tăng dài nhất tại vị trí i
    L = [1] * n

    for i in range(n):
        for j in range(i):
            if arr[i] > arr[j]:
                L[i] = max(L[i], L[j] + 1)
    return max(L)

arr = eval(input())

print(longest_sequence(arr))
```

```python
#W7A8
import bisect

def solve(intervals, queries):
    #ý tưởng giải bài này: với mỗi một intervals (từ ngắn đến dài), tìm các q truy vấn nằm trong khoảng interval đó và cập nhật res
    
    res = [-1] * len(queries)

    #sort các intervals theo độ dài, xử lý các khoảng ngắn nhất trước
    intervals.sort(key = lambda x : x[1] - x[0] + 1)

    #để dùng được binary search trên queries, ta phải sort, nhưng khi sort thì sẽ bị mất thứ tự ban đầu, nên phải tạo mảng lưu index ban đầu

    queries_with_idx = sorted(list(enumerate(queries, 0)), key = lambda x : x[1])

    sorted_queries = [x[1] for x in queries_with_idx]

    original_idx = [x[0] for x in queries_with_idx]

    #xem cái interval có thể chứa những queries nào
    for begin , end in intervals:
        length = end - begin + 1

        left = bisect.bisect_left(sorted_queries, begin)

        right = bisect.bisect_right(sorted_queries, end)

        for i in range(left, right):

            if res[original_idx[i]] == -1:

                res[original_idx[i]] = length

    return res

intervals = eval(input())

queries = eval(input())

print(solve(intervals, queries))


```

```python
#W7A9

def solve(nums):
    res = []
    n = len(nums)
    for i in range(n):
        cnt = 0
        for j in range(i + 1, n):
            if nums[j] < nums[i]:
                cnt += 1
        res.append(cnt)
    return res

nums = eval(input())

print(solve(nums))
```

```python
#W7A15
n = int(input())

names = [input() for _ in range(n)]

def find_nemo(names):
    for i in range(len(names)):
        if names[i] == "Nemo":
            return i
        
index = find_nemo(names)

if (index - 1) == -1:
    print(names[-1], end = " ") 
    print("and", end = " ")

else:
    print(names[index - 1], end = " ")
    print("and", end = " ")

if (index + 1) == len(names):
    print(names[0])

else:
    print(names[index + 1])    
```
